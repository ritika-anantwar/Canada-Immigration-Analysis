# Visualizing Data using Matplotlib

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

# Line plot showing immigration from Haiti

```python
haiti = df.loc['Haiti', years]
haiti.head()
```

```python
#haiti.plot()

#OR

haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
The spike seen in 2010 was due to an earthquake in Haiti.
```

## Comparing the trend of top 5 countries that contributed the most to immigration to Canada.

```python
inplace=True
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)

top5 = df.head(5)
top5 = top5[years].transpose()
top5
```

## Line plot:-

```python
top5.index = top5.index.map(int)
top5.plot(kind='line', figsize = (14,8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

## Area plot:-

```python
top5.index = top5.index.map(int)
top5.plot(kind='area', stacked=False, figsize = (20, 10), alpha=0.25)

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

## Histogram:-

```python
df['2013'].head()
```

```python
count, bin_edges = np.histogram(df['2013'])
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind='hist', figsize=(8, 5), xticks = bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
#Immigration distribution for Denmark, Norway and Sweden

df.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_DNS = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_DNS.head()
```

```python
count, bin_edges = np.histogram(df_DNS, 15)

df_DNS.plot(kind='hist', figsize=(10, 6), xticks=bin_edges, color=['coral', 'darkslateblue', 'mediumseagreen'], bins=15)

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

## Barplot:-

```python
#Immigrants from Iceland in 1980-2013
df_iceland = df.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

plt.annotate('', xy=(32, 70), xytext=(28, 20), xycoords='data', arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))

plt.annotate('2008 - 2011 Financial Crisis', xy=(28, 30), rotation=72.5, 
             va='bottom', ha='left')
plt.show()
```

```python
df.sort_values(by='Total', ascending=False, inplace=True)
df_top15 = df['Total'].head(15)
df_top15

df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

for index, value in enumerate(df_top15): 
    label = format(int(value), ',')
    
plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')
plt.show()
```
